# **One Hot Encoding For Multi Categorical Variables**

kaggle dataset:mercedsbenz.csv

**load categorical variables**

```python
import numpy as np
import pandas as pd

#load categorical variables of dataset

data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

**How many unique categories in each and every column**

```python
#unique label in each column variables
for col in data.columns:
  print(col,': ',len(data[col].unique()),'labels')
```

**Lets check how many columns will be generated after one hot encoding to 6 variables**

```python
pd.get_dummies(data,drop_first=True).shape
```

just from 6 initial categorical variables we generate 117 new variables


As number of columns/variables increase,it leads to curse of dimensionality whichh will impact acuracy


**SOLUTION**:

Take **most frequent** 10 categories/**label**s from each features nad skip the rest from each column.

Apply one hot encoding to those 10 frequent labels of the variable.

1 Binary variable for each of 10 frequent labels of the variable only.

Group all the rest under a new category and drop 

Finding topmost 20 values/categories of  X2 variable

```python
 data.X2.value_counts().sort_values(ascending=False).head(20)
```

Lets take top 10  frequent categories of the variable X2 from previous step as a list

```python
top_10_labels=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
# index edukumbol we get the categories name
top_10_labels
```

Next we will make binary variables

ensemble technique

```python
for label in top_10_labels:
  data[label]=np.where(data['X2']==label,1,0)
  #label indel replace 1 else 0
data[['X2']+top_10_labels].head(40)
#idhupole create matrix for all categories
```

```python
# get whole set of dummy variables, for all the categorical variables

def one_hot_encoding_top_x(df, variable, top_x_labels):
  #variable#column name,
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label, 1, 0)
```

```python
# read the data again
df = pd.read_csv('mercedesbenz.csv', usecols=['X1', 'X2'])

# encode X2 into the 10 most frequent categories
one_hot_encoding_top_x(df, 'X2', top_10_labels)
df.head()
```

# **Advantages of OHE**



1.   Straightforward ti implement and doesnot require hours of variable exploration
2.   Doesnot expand massively the feature space



# **Disadvantages of OHE**


1.   Doesnot add info that may make the variable more predictive
2.   Doesnot keep info of ignored labels



So here we consider dominating categorical variables and considered the minorities as noise
